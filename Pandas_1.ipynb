```python
# Creating your own dataframe.
# Number entries

test_scores = pd.DataFrame({'Student_ID': [154,973,645], 
                            'Science': [50,75,31], 'Geography': [88,100,66],
                           'Math': [72,86,94]})
test_scores
```

```python
# Text entries
survey = pd.DataFrame({'James': ['I liked it','more salt'],
                      'Emily': ['It is too sweet','Yum!']})
survey
```

```python
# Access of Student's data from his roll_no. 
```

```python
# Index -- We can either set an existing column as our index or specify an index when creating a dataframe.
# Let's begin by setting an existing column as index.

test_scores = test_scores.set_index('Student_ID')
test_scores
```

```python
survey = pd.DataFrame({'James':['I liked it', 'more salt'], 'Emily': ['It is too sweet','Yum!']},
                     index = ['Product A', 'Product B'])
survey
```

```python
# You can also reset the index back to its default.
# Reset Index
# Try playing around with 'drop' and 'inplace' and see what they do.

survey.reset_index(drop=True,inplace=True)

# inplace=True -- means to save the changes.
```

```python
survey
```

```python
# Renaming we want to change the names of the first two columns.
test_scores.rename(columns = {'Geography': 'Physics', 'Science': 'Arts'}, inplace = True)
test_scores
```

```python
# Drop the 'Math' column
test_scores.drop(columns='Math',inplace=True)
test_scores
```

```python
# Drop row with student_ID 973
# We can make this more robust once we learn the 'loc' function in the coming weeks
test_scores.drop(973,inplace=True)
test_scores
```
