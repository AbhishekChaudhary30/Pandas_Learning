```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('replace_data.csv')
df
```

```python
# Replacing single value.
new_df = df.replace(-99999, value=np.nan) # null " "
new_df
```

```python
df = pd.read_csv("replace_data1.csv")
df
```

```python
# Replacing multiple values.
new_df = df.replace([-99999, -88888, "no event"], np.nan)
new_df
```

```python
df = pd.read_csv('replace_data1_1.csv')
df
```

```python
# Replacing per column
new_df = df.replace({'temperature': -99999, 'windspeed': -88888, 'event': 'no event'}, np.nan)
new_df
```

```python
# Replacing by using mapping
df = pd.read_csv('replace_data1.csv')
df
```

```python
new_df = df.replace({-99999: np.nan, 'no event': 'Sunny'})
new_df

new_df.to_csv('file_name.csv')
# csv, excel
# tsv,txt
```

```python
df1 = pd.DataFrame({'Name': ['Vicky', 'Bill'], 'Age': [12, 46]})
df2 = pd.DataFrame({'Name': ['John', 'Sabrina'], 'Age': [37,25]})
df1
```

```python
df2
```

```python
# Concat horizontally
cars = pd.concat([df1, df2], axis=1)
cars
```

```python
# concat vertically and ignore index
df3 = pd.concat([df1, df2], axis=0, ignore_index=True)
df3
```

```python
df4 = pd.DataFrame({'Name': ['Tyler', 'Natalie'], 'Age': [28,39], 'Hobby': ['Swimming', 'Reading']})
df4
```

```python
# Outer is the default for the join argument --- Outer does not consider any difference in column b/w two dataframes.
pd.concat([df3, df4], join = 'outer', ignore_index=True)
```

```python
# Inner stacks only the columns that are shared between the two dataframes.
pd.concat([df3,df4], join = 'inner', ignore_index=True)
```

```python
# Merge --  focus of rows.
size = pd.DataFrame({'Colour': ['Red', 'Blue', 'Green'], 'Size': ['M', 'S', 'M']})
gender = pd.DataFrame({'Colour': ['Red', 'Blue', 'Yellow'], 'Sex': ['Female', 'Female', 'Male']})
size
```

```python
gender
```

```python
pd.merge(size, gender, how='inner')
```

```python
pd.merge(size, gender, how='outer')
```

```python
pd.merge(size, gender, how='left')
```

```python
pd.merge(size, gender, how='right')
```
