```python
import pandas as pd
```

```python
data = pd.read_csv('employee.csv')
data.head()
```

```python
# GroupBy
# univariate analysis -- single column study
# bivariate analysis -- two columns study
# multivariate analysis -- more than two columns
```

```python
data.columns
```

```python
data["Attrition"].unique()
```

```python
data["Attrition"].value_counts() # mode
```

```python
data["Age"].mean()
```

```python
data["Age"].median()
```

```python
# return the mean age of employees according to their department.
# syntax - data.group(categorical)[Numerical].mean(), max, median
data.groupby("Department")["Age"].mean()
```

```python
# # return the mean age of employees according to their EducationField.
data.groupby("EducationField")["Age"].mean()
```

```python
# return the mean of MonthlyIncome according to gender.
data.groupby("Gender")["MonthlyIncome"].mean()
```

```python
data.groupby("Gender")["EducationField"].value_counts()
```

```python
# Multivariate -----
# mean TotalWorkingYears with respect to Gender and Department
data.groupby(["Gender", "Department"])["TotalWorkingYears"].mean()
```

```python
# # mean TotalWorkingYears, MonthlyIncome with respect to Gender and Department
data.groupby("Gender")[["TotalWorkingYears","MonthlyIncome"]].mean()
```
