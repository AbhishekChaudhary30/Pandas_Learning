```python
import pandas as pd
```

```python
train = pd.read_csv("titanic_train.csv") # here train is a variable.
train.head()
```

```python
train.describe() # shows the stats of all numerical column
               # train.describe(include ="all")
```

```python
train.describe(include ="all")
```

```python
# Describe function on numerical variable
train["Fare"].describe()
```

```python
# Describe function on text variable
train['Embarked'].describe() # train["Embarked"].value_counts()
```

```python
# Unique and value counts functions
# How many unique Embarked values are there?

train['Embarked'].nunique() # n is number of uniques
```

```python
# What are the unique Embarked values?
train['Embarked'].unique()
```

```python
# Descriptive statistics

# what is the oldest age?
train['Age'].max()
```

```python
# who is that passenger ---- 
train.loc[train['Age'] == 80.0, :]
```

```python
# What is the youngest age?
train['Age'].min()
```

```python
# Who is that passenger ? 
train.loc[train['Age'] == 0.42, :]
```

```python
# What is the average age?
train['Age'].mean()
```

```python
# What is the median?
train['Fare'].median()
```

```python
# What is the most frequent Embarked value?

train['Embarked'].mode()
```
